```python
#importing libraries 

import pandas as pd
import json 
```

**Process Match result**

```python
#opning match result file
with open('D:/harshita/project/file2/t20_json_files/match_results.json') as f:
    data = json.load(f)

#data[0]['matchSummary'] 

#convert it into datafram
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match
```

**Renaming scorecard column as match id to link with other tables**

```python
#renaming the column name scorecard to match_id
df_match.rename({'scorecard':'match_id'},axis= 1,inplace=True)
```

**create a match dictionary that maps team names to unique match id. this will be helpfull later to link with other table**

```python

match_dict={}
for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_dict[key1]= row['match_id']
    match_dict[key2]= row['match_id']
match_dict    
```

```python
df_match.to_csv('D:/harshita/project/file2/csv/match.csv', index = False)
```

**process batting summary**

```python
#opening file 'batting summary'

with open('D:/harshita/project/file2/t20_json_files/batting_summary.json') as f:
    data = json.load(f)   

#collecting all data together
all_records=[]
for rec in data:
    all_records.extend(rec['battingSummary'])

#creating dataframe
df_batting = pd.DataFrame(all_records)

df_batting.head(5)
```

```python
#adding new column 'out or not-out'
df_batting['out/not_out']= df_batting.dismissal.apply(lambda x: 'out' if len(x)>0 else 'not-out' )
df_batting.head(5)
```

```python
#create a new column called 'match_id'
df_batting['match_id'] = df_batting['match'].map(match_dict)
df_batting.head(5)
```

```python
#delete the column named 'dismissal'
df_batting.drop(columns= ["dismissal"],inplace= True)
df_batting.head(10)
```

```python
df_batting.head(50)
```

**clean weird characters**

```python
#modifications in batsmanname
df_batting['batsmanName']= df_batting['batsmanName'].apply(lambda x : x.replace('â€',' '))

df_batting.head(50)
```

```python
df_batting['batsmanName']= df_batting['batsmanName'].apply(lambda x : x.replace('(c)',' '))

df_batting.head(50)
```

```python
df_batting.to_csv('D:/harshita/project/file2/csv/batting_summary.csv',index=False)
```
